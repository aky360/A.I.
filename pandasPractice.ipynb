```python
import pandas as pd
```

```python
df = pd.DataFrame([[1, 2, 3], [None, None, 6], [None, None, 9]])
df
```

```python
df[1].fillna(method='ffill', inplace=True)
df
```

```python
df1 = pd.DataFrame([[1, 2, 3], [None, None, 6], [7, None, 9]])
df1
```

```python
df1[0].fillna(method='bfill', inplace=True)
df1
```

```python
df1[1].fillna(method='ffill', limit=1, inplace=True)
df1
```

```python
df2 = pd.DataFrame([[1, 2, 3], [None, None, 6], [None, None, 9]])
df2
```
