```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/aky360/A.I./main/Data%20Science/Data%20Analysis/Datasets/6.%20India%20Census%202011.csv"
df = pd.read_csv(url)
df
```

```python
print(df.shape)
print(df.size)
print(df.index)
print(df.columns)
print(df.columns.size)
print(df.dtypes)
```

```python
df.describe()
```

```python
df.iloc[0:6,:].style
```

```python
df.iloc[0:6,:].style.hide_index()
```

```python
df.iloc[0:6,:].style.set_caption('India Census 2011 Datasets')
```

```python
df[df['District_name'].isin(['New Delhi', 'Lucknow', 'Jaipur'])].style
```

```python
df.groupby('State_name')['Population'].sum()
```

```python
df.groupby('State_name')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum()
```

```python
df.groupby(df['State_name'] == 'MAHARASHTRA')['Male_Workers'].sum()[True]
```

```python
df[df['State_name'] == 'MAHARASHTRA']['Male_Workers'].sum()
```
