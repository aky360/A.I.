**Q. 1) Instruction ( For Data Cleaning ) - Remove the column that only contains missing values.**

---

```python
df.drop(columns = 'country_name', inplace=True)
```

```python
df.head(3)
```

**Q. 2) Question ( Based on Filtering + Value Counts ) - For Speeding , were Men or Women stopped more often ?**

---

```python
df[df['violation'] == 'Speeding']
```

```python
df[df['violation'] == 'Speeding']['driver_gender']
```

```python
df[df['violation'] == 'Speeding']['driver_gender'].value_counts()
```

```python
df.groupby('driver_gender')['violation'].value_counts()
```

```python
# genders = df.groupby('driver_gender')['violation'].value_counts()
# print(type(genders))
df.groupby(['driver_gender', 'driver_race'])['violation'].value_counts()
```

**Q. 3) Question ( Groupby ) - Does gender affect who gets searched during a stop ? Question ( mapping + data-type casting )**


---

```python
df[df['search_conducted'] == True]['driver_gender'].value_counts()
```

```python
df.groupby('driver_gender')['search_conducted'].sum()
```

```python
df['search_conducted'].value_counts()
```

**Q. 4) Question ( mapping + data-type casting ) - What is the mean stop_duration ?**

---

```python
df['stop_duration'].value_counts()
```

```python
df['stop_duration'] = df['stop_duration'].map({'0-15 Min' : 7.5, '16-30 Min' : 24, '30+ Min' : 45})
```

```python
df.head(3)
```

```python
df['stop_duration'].mean()
```
